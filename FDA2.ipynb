```python
import pandas as pd
import numpy as np
# Defined 4 things in the global variable. 3 arrays T,C,S and option that is the outputs when the dice is rolled
# The outputs when Coin is filped H or T
def CT(n):
  op = ['H','T']
  T = []
  C= []
  S = []
# since they have mentioned 1 <= n <= 3 
# so there is only 3 test onditions
# they are when n = 1, n= 2 , n = 3 

# Used 3 if conditions to 

#Used a for loop to iterate through and get the required output  and appended to an array 

  if n ==1:
    x = [1,2]
    for i in op[0:2]:
      T.append(i)
#1st dataframe is used to print the index value 
# 2nd dataframe is used to print the no of outcomes when 1 coins are fliped   
    df1 = pd.DataFrame(np.array(x), columns=[ 'Number'])
    df2 = pd.DataFrame(np.array(T), columns=[ 'Toss 1'])
    dfs = [df1 ,df2]
    finals = pd.concat(dfs, axis=1)
    finals.set_index('Number', inplace=True)
    return finals
#Used 2 for loops to iterate through and appended into a array so that we can put it in dataframe  
  if n ==2 :
    x = [1,2,3,4]
    for i in op[0:2]:
      for j in op[0:2]:
        T.append(i)
        C.append(j)
#1st dataframe is used to print the index value 
# 2nd , 3rd  dataframe is used to print the no of outcomes when 2 coins are fliped    
    df = pd.DataFrame(np.array(x), columns=[ 'Number'])
    df1 = pd.DataFrame(np.array(T), columns=[ 'Toss 1'])
    df2 = pd.DataFrame(np.array(C), columns=[ 'Toss 2'])
    dfs = [ df , df1 ,df2]
    finals = pd.concat(dfs, axis=1)
    finals.set_index('Number', inplace=True)
    return finals

#Used 3 for loops to iterate through and appended into a array so that we can put it in dataframe  
  if n ==3 :
    x = [1,2,3,4,5,6,7,8]
    for i in op[0:2]:
      for j in op[0:2]:
        for k in op[0:2]:
          T.append(i)
          C.append(j)
          S.append(k)
    
    df = pd.DataFrame(np.array(x), columns=[ 'Number'])
    #1st dataframe is used to print the index value 
# 2nd , 3rd  and 4th dataframe is used to print the no of outcomes when 3 coins are fliped
    df1 = pd.DataFrame(np.array(T), columns=[ 'Toss 1'])
    df2 = pd.DataFrame(np.array(C), columns=[ 'Toss 2'])
    df3 = pd.DataFrame(np.array(S), columns=[ 'Toss 3'])
    dfs = [ df , df1 ,df2 , df3]
    finals = pd.concat(dfs, axis=1)
    finals.set_index('Number', inplace=True)
    return finals
# Since the condition should be 1 <= N <= 3
#if any other number is entered it shows wrong output
  print('Wrong Input')    
  print('Enter number between 1 to 3')




```

**4.Write custom functions to generate sample space and counting**

```python

def ss(n, r, ordered, replacement):
  if n == 0 and r == 0 :
    print("Wrong Input")

# n is the number of objects in the box
# r is the number of objects selected
# Ordered is a boolean value (true or false)
# Replacement is true or false (true or false)

#When order is true and replacement true 
#We have to print all kinda combination when 2 elements is choosen out of 3 

#When the order is true and replacement is true 
  if ordered == True and replacement == True :
    num = []
    X1 = []
    X2 = []
    a = pow(n,r)
    for o in range(1,a+1):
      num.append(o)
    for i in range(1,n+1):     # iterated using for loop 
      for j in range(1,n+1):    
        X1.append([i,j])       #appended the every iteration to a loop 
    df = pd.DataFrame(np.array(num), columns=[ ''])
    df2 = pd.DataFrame(np.array(X1), columns=[ 'X1' , 'X2'])
    dfs = [df ,df2 ]
    final = pd.concat(dfs, axis=1)
    final.set_index('', inplace=True)
    return final

#When the order is true and replacement is false 
  if ordered == True and replacement == False :
    num = []
    X1 = []
    X2 = []
    a = pow(n,r) - n 
    for o in range(1,a+1): #iteration for number 
      num.append(o)
    for i in range(1,n+1):   # iterated using for loop
      for j in range(1,n+1):
        if i != j:
          X1.append([i,j])
    df = pd.DataFrame(np.array(num), columns=[ ''])
#1st dataframe is used to print the index value 
# 2nd  dataframe is used to print the no of outcomes 
    df2 = pd.DataFrame(np.array(X1), columns=[ 'X1' , 'X2'])
    dfs = [df ,df2 ]
    final = pd.concat(dfs, axis=1)
    final.dropna(inplace=True)
    final.set_index('', inplace=True)
    return final

#When the order is False and replacement is true 
  if ordered == False and replacement == True :
    num = []
    X1 = []
    X2 = []
    a = pow(n,r) - n 
    for o in range(1,a+1):
      num.append(o)
    for i in range(1,n+1):  # iterated using for loop
      for j in range(1,n+1):
        if i <= j:
          X1.append([i,j])  #appended the every iteration to a loop 
    df = pd.DataFrame(np.array(num), columns=[ ''])
    #1st dataframe is used to print the index value 
# 2nd  dataframe is used to print the no of outcomes 
    df2 = pd.DataFrame(np.array(X1), columns=[ 'X1' , 'X2'])
    dfs = [df ,df2 ]
    final = pd.concat(dfs, axis=1)
    final.dropna(inplace=True)
    final.set_index('', inplace=True)
    return final

#When the order is False and replacement is False 
  if ordered == False and replacement == False :
    num = []
    X1 = []
    X2 = []
    a = n
    for o in range(1,a+1):
      num.append(o)
    for i in range(1,n+1):
      for j in range(1,n+1): # iterated using for loop
        if i <= j and i != j:
          X1.append([i,j])  #appended the every iteration to a loop 
    df = pd.DataFrame(np.array(num), columns=[ ''])
    #1st dataframe is used to print the index value 
# 2nd  dataframe is used to print the no of outcomes 
    df2 = pd.DataFrame(np.array(X1), columns=[ 'X1' , 'X2'])
    dfs = [df ,df2 ]
    final = pd.concat(dfs, axis=1)
    final.set_index('', inplace=True)
    return final


      
```
